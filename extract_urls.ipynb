```python
# Define the input and output file paths
input_file_path = "urls.txt"
output_file_path = "unique_urls.txt"

# Initialize a set to store unique URLs
unique_urls = set()

# Read the input file and add unique URLs to the set
with open(input_file_path, "r") as input_file:
    for line in input_file:
        url = line.strip()  # Remove leading/trailing whitespace
        if url:  # Check if the line is not empty
            unique_urls.add(url)

# Write the unique URLs to the output file
with open(output_file_path, "w") as output_file:
    for url in unique_urls:
        output_file.write(url + "\n")

print(f"Unique URLs saved to {output_file_path}")
```
